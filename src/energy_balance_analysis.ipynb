# 3. Analiza energetske bilance

**Pri tem vprašanju nas je zanimala analiza energetske bilance skozi leta in sprememba porabe energije glede na panogo.**

#### 3.1 Pregled podatkov v potrebnih podatkovnih množicah

Najprej smo pregledali podatke v podatkovnih množicah, ki smo jih uporabljali. To smo naredili z izpisom glave podatkovnega okvirja.

```python
from reading import data_extraction as de
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import misc_utils as mu

energy_balance = de.read('../data/Energetska bilanca.csv')
energy_use = de.read('../data/Energetska poraba goriv, električne energije in toplotne energije po dejavnosti.csv')

mu.display_side_by_side(energy_balance.head(), energy_use.head(), titles=['Energetska bilanca', 'Energetska poraba'], suppres_scientific=True)
```

Kot lahko vidimo, so podatkovne množice v precej neuporabni obliki, saj želimo, da so v stolpcih samo leta, ostali atributi pa so v posebnem stolpcu. Tak format želimo zato, da lahko podatke obdelujemo po časovnih obdobjih, brez da bi bilo potrebno manipuliranje imen stolpcev.

#### 3.2 Preoblikovanje podatkovnih množic

Najprej smo se lotili preoblikovanja podatkovnih množic v format, ki je bolj primeren za našo uporabo.

```python
# Get unique values of energy sources and years
balance_sources = list(dict.fromkeys([' '.join(col.split()[1:]) for col in energy_balance.columns[1:]]))
balance_years = sorted(list(set([col.split()[0] for col in energy_balance.columns[1:]])))

# Create new dataframe
balance = pd.DataFrame()

# Add needed columns to dataframe
balance.insert(0, 'OSKRBA IN PORABA', '')
balance.insert(1, 'ENERGETSKI VIR', '')
count = 2
for year in balance_years:
    balance.insert(count, year, np.nan)
    count += 1

## Fill dataframe with data
# Add first two columns
for oip in energy_balance['OSKRBA IN PORABA']:
    for source in balance_sources:
        balance.loc[len(balance.index)] = [oip, source] + [np.nan] * len(balance_years)

# Add values from Energetska bilanca.csv
for col in energy_balance.columns[1:]:
    # Extract year and energy source from column name
    year = col.split()[0]
    source = ' '.join(col.split()[1:])
    # Loop throug all rows in dataframe and add value to appropriate cell
    for index, row in energy_balance.iterrows():
        if row[col] != '-':
            balance.loc[(balance['OSKRBA IN PORABA'] == row['OSKRBA IN PORABA']) & (balance['ENERGETSKI VIR'] == source), year] = row[col]


mu.display_side_by_side(balance.head())
```

```python
# Get unique values of energy sources and years
use_sources = list(dict.fromkeys([' '.join(col.split()[7:]).capitalize() for col in energy_use.columns[1:]]))
use_years = sorted(list(set([col.split()[0] for col in energy_use.columns[1:]])))

# Create new dataframe
use = pd.DataFrame()

# Add needed columns to dataframe
use.insert(0, 'PODROČJA IN PODPODROČJA DEJAVNOSTI', '')
use.insert(1, 'ENERGETSKI VIR', '')
count = 2
for year in use_years:
    use.insert(count, year, np.nan)
    count += 1	

## Fill dataframe with data
# Add first two columns
for field in energy_use['PODROČJA IN PODPODROČJA DEJAVNOSTI']:
    for source in use_sources:
        use.loc[len(use.index)] = [field, source] + [np.nan] * len(use_years)
        
# Add values from Energetska poraba goriv, električne energije in toplotne energije po dejavnosti.csv
for col in energy_use.columns[1:]:
    # Extract year and energy source from column name
    year = col.split()[0]
    source = ' '.join(col.split()[7:]).capitalize()
    # Loop throug all rows in dataframe and add value to appropriate cell
    for index, row in energy_use.iterrows():
        if row[col] != '-':
            use.loc[(use['PODROČJA IN PODPODROČJA DEJAVNOSTI'] == row['PODROČJA IN PODPODROČJA DEJAVNOSTI']) & (use['ENERGETSKI VIR'] == source), year] = row[col]
            
mu.display_side_by_side(use.head())
```

Podatkovne množice smo preoblikovali tako, da smo ustvarili nova podatkovna okvirja, katerima smo pripisali potrebne stolpce, nato pa smo v ustrezne celice vstavili vrednost iz originalnih podatkovnih množic. Tako smo dobili podatkovne okvirje, s katerimi je lažje delati (ekstrahirati vrednosti, analizirati vrednosti v času, itd.).

#### 3.3 Analiza energetske bilance

Pri tem podvprašanju smo analizirali energetsko bilanco skozi čas, bolj specifično, pogledali smo skupno končno porabo energetskih virov v letnih intervalih, analizirali smo porabo energentov po sektorjih, ki so se nam zdeli najbolj zanimivi, in analizirali uvoz/izvoz naftnih proizvodov, saj so najbolj porabljan energent.

```python
# Create dictionary with energy sources
final_usage = mu.init_dict_based_on_column(balance, 'ENERGETSKI VIR')

# Fill dictionary
for year in balance.columns[2:]:
    for key in final_usage.keys():
        final_usage[key][year] = balance[year][balance['ENERGETSKI VIR'] == key][balance['OSKRBA IN PORABA'] == 'Končna poraba'].to_list()[0]

# Flatten internal dict
final_usage_f = mu.flatten_internal_dict(final_usage)

years = balance.columns[2:]

# Plot different energy sources
plt.figure(figsize=(20, 5))
plt.suptitle("Končna poraba energetskih virov", fontsize=20)
plt.xticks(rotation=90)
plt.ylabel("Poraba energenta (TJ)")
plt.grid()
for key, value in final_usage_f.items():
    if key != 'Energetski viri-SKUPAJ':
        plt.plot(years, value, label=key)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
print()
```

Opazili smo, da vsi sektorji skupaj največ porabijo naftnih proizvodov, takoj za tem je električna energija, zemeljski plin ter obnovljivi viri. Posebej zanimiv je padec v porabi nekaterih energentov (naftni proizvodi, električna energija, zemeljski plin) v letu 2009, in močan padec v porabi naftnih proizvodov iz leta 2019 na 2020.

Predvidevamo, da je do padca porabe naftnih proizvodov leta 2020 prišo zaradi prvega vala COVID-19, saj je bila država zaprta, in zaradi tega ljudje niso vozili, tovarne niso obratovale, itd.

Kar se tiče vpadanja porabe nekaterih energetskih virov leta 2009, to potrjuje tudi [Direktorat za energijo, str. 9](http://www.energetika-portal.si/fileadmin/dokumenti/publikacije/energija_slo/evs_2009.pdf), ki pravi, da je bila oskrba z energijo leta 2009 v primerjavi z 2008 manjša za 8.5%, iz domačih virov je bilo pridobljeno 0.3% manj energije, uvoz pa je bil manjši za celih 12.1%. Predvidevamo, da je do tega prišlo zaradi globalne krize leta 2009, ki je močno udarila na mnoge sektorje, ki porabljajo navedene energente.

##### 3.3.1 Analiza porabe po nekaterih sektorjih (kmetijski, prometni, neenergetski)

```python
# Create dicts of energy sources and their usages for each sector (farming, traffic, non-energy) in balance dataframe
balance_farming = mu.init_dict_based_on_column(balance, 'ENERGETSKI VIR')
balance_traffic = mu.init_dict_based_on_column(balance, 'ENERGETSKI VIR')
balance_non_energy = mu.init_dict_based_on_column(balance, 'ENERGETSKI VIR')

# Fill dicts with data
for year in balance.columns[2:]:
    for key in balance_farming.keys():
        balance_farming[key][year] = balance[year][balance['ENERGETSKI VIR'] == key][balance['OSKRBA IN PORABA'] == 'Končna poraba-Kmetijstvo in gozdarstvo'].to_list()[0]
        balance_traffic[key][year] = balance[year][balance['ENERGETSKI VIR'] == key][balance['OSKRBA IN PORABA'] == 'Končna poraba-Promet'].to_list()[0]
        balance_non_energy[key][year] = balance[year][balance['ENERGETSKI VIR'] == key][balance['OSKRBA IN PORABA'] == 'Končna poraba-Neenergetska raba'].to_list()[0]

balance_farming_f = mu.flatten_internal_dict(balance_farming)
balance_traffic_f = mu.flatten_internal_dict(balance_traffic)
balance_non_energy_f = mu.flatten_internal_dict(balance_non_energy)
```

```python
# Display plots of energy use
plt.figure(figsize=(30, 5))
ax = plt.subplot(1, 3, 1)
ax.set_title('Kmetijstvo in gozdarstvo')
plt.suptitle("Poraba energetskih virov", fontsize=20)
plt.xticks(rotation=90)
plt.ylabel("Poraba energenta (TJ)")
plt.grid()
for key, value in balance_farming_f.items():
    if key != 'Energetski viri-SKUPAJ':
        plt.plot(years, value, label=key)
plt.legend()

ax = plt.subplot(1, 3, 2)
ax.set_title('Promet')
plt.xticks(rotation=90)
plt.ylabel("Poraba energenta (TJ)")
plt.grid()
for key, value in balance_traffic_f.items():
    if key != 'Energetski viri-SKUPAJ':
        plt.plot(years, value, label=key)
plt.legend()

ax = plt.subplot(1, 3, 3)
ax.set_title('Neenergetska raba')
plt.xticks(rotation=90)
plt.ylabel("Poraba energenta (TJ)")
plt.grid()
for key, value in balance_non_energy_f.items():
    if key != 'Energetski viri-SKUPAJ':
        plt.plot(years, value, label=key)
plt.legend()
print()
```

Pri analizi porabe energentov v različnih sektorjih, smo po pričakovanjih opazili da kmetijstvo in gozdarstvo ter promet porabita največ naftnih proizvodov, kot so motorno gorivo in dizelsko gorivo, o katerih smo govorili že v prejšnjem vprašanju. 

Zanimiv je porast porabe geotermalne in sončne energije v kmetijstvu in gozdarstvu, ki se je začel leta 2009, sovpada pa z padcem porabe naftnih proizvodov. Predvidevamo, da je do tega prišlo zaradi globalne krize, in potrebe po osamosvojitvi oz. zmanjšanju odvisnosti od tujih virov. 

Poleg tega je zanimiva tudi majhna poraba obnovljivih virov v prometnem sektorju, ki se je začela leta 2007. To potrjuje tudi [European Environment Agency](https://www.eea.europa.eu/ims/use-of-renewable-energy-for), ki pravi, da je leta 2009 Renewable Energy Directive postavil tarče za države članice, da mora do leta 2020, vsaj 10% energentov porabljenih v transportu prihajati iz obnovljivih virov energije.

Kar pa se tiče [neenergetskega sektorja](http://www.interenerstat.org/definitions/results.asp?id=168&Type=Flows), opazimo, da se uporabljajo le trdna goriva, naftni proizvodi in zemeljski plin. Zanimivo je, da je z leta 2009 na leto 2011 poraba zemeljskega plina vpadla skoraj na 0TJ, in da od takrat zelo počasi narašča. Prav tako je v tem časovnem obdobju rahlo upadla uporaba trdnih energentov in naftnih proizvodov. Predvidevamo, da je do tega prišlo zaradi takratne globalne krize, ki je vplivala na delovanja raznih tovarn, in pa napredkov na samem področju, ki so facilitirali uporabo drugih virov.

##### 3.3.2 Analiza uvoza/izvoza naftnih proizvodov

```python
oil_products = dict.fromkeys(['Domača proizvodnja', 'Uvoz', 'Izvoz', 'Spremembe zalog'])

for key in oil_products.keys():
    oil_products[key] = dict.fromkeys(balance.columns[2:])
    for year in balance.columns[2:]:
        oil_products[key][year] = balance[year][balance['OSKRBA IN PORABA'] == key][balance['ENERGETSKI VIR'] == 'Naftni proizvodi'].to_list()[0]
        
oil_products_f = mu.flatten_internal_dict(oil_products)

# Display plots of energy use
plt.figure(figsize=(20, 5))
plt.suptitle("Analiza naftnih proizvodov", fontsize=20)
plt.xticks(rotation=90)
plt.ylabel("Količina energenta (TJ)")
plt.grid()
for key, value in oil_products_f.items():
    plt.plot(years, value, label=key)
plt.legend()
print()
```

Iz prikazanega grafa smo opazili, da v Sloveniji nikoli nismo proizvajali naftnih proizvodov, kar je smiselno, saj nimamo naravnih rezursov za to.
